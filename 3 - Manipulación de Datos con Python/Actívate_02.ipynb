```python
from pickle import TRUE
new_row1=pd.Series(["Juan","Frausto","Male","México",25,'14/05/1997',78895])
new_row2=pd.Series(["Max","Arias","Male","México",21,'19/11/1999',45612])
new_row3=pd.Series(["Norma","Avila","Female","México",50,'14/12/1940',12345])
new_rows_df = pd.DataFrame(new_row1)
new_rows_df = pd.concat([new_rows_df, new_row2], axis=1)
new_rows_df = pd.concat([new_rows_df, new_row3], axis=1)
new_rows_df = new_rows_df.T
new_rows_df.columns = ["First Name",'Last Name', 'Gender', 'Country', 'Age', 'Date', 'Id']
#new_rows_df.reset_index(drop = TRUE)
new_rows_df
```
