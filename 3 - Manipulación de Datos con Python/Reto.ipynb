```python
datos2 = train.merge(train2, left_on='Employee_ID', right_on='Employee_ID', how = 'right')
datos2
```

```python
datos1 = test.merge(test2, left_on='Employee_ID', right_on='Employee_ID', how = 'right')
datos1
```

```python
datos=pd.concat([datos1,datos2], axis = 0)
datos
```

```python
datos.rename(columns={' Computer_OS':'Computer_OS', ' Mobile_OS':'Mobile_OS', ' Mobile_tickets':'Mobile_tickets', ' Computer_tickets':'Computer_tickets'}, inplace=True)
```

```python
#datos = datos.replace({'Computer_OS':'No registradoa'},"No registrado")
#datos = datos.replace({'Mobile_OS':np.nan},"No registrado")
datos["Computer_OS"] = datos["Computer_OS"].replace(np.nan, 'No registrado')
datos["Mobile_OS"] = datos["Mobile_OS"].replace(np.nan, 'No registrado')
datos.to_string()
```

```python
print(datos.columns.tolist())
```

### El porcentaje de personas que usan Windows / Mac / Linux / no registrado como su sistema operativo en su computadora 

```python
freq1 = datos['Computer_OS'].value_counts() 
total1 = datos['Computer_OS'].count()
resul = (freq1/total1)*100
resul
```

### El porcentaje de personas que usan iOS / Android / No registrado como su sistema operativo en su dispositivo móvil

```python
freq2 = datos['Mobile_OS'].value_counts() 
total2 = datos['Mobile_OS'].count()
resul2 = (freq2/total2)*100
resul2
#datos['Computer_OS'] = (datos['Computer_OS'] / datos['Computer_OS'].sum()) * 100
```

### La edad promedio de la gente que utiliza cada sistema operativo de escritorio

```python
freq = datos.groupby(['Computer_OS']).mean() 
resul3 = freq['Age']
resul3
```

### El nivel educativo promedio de la gente que utiliza cada sistema operativo de escritorio

```python
freq = datos.groupby(['Computer_OS']).mean() 
resul4 = freq['Education_Level']
resul4
```

### El número total de tickets generados por la gente con cada sistema operativo de escritorio

```python
freq = datos.groupby(['Computer_OS']).sum() 
resul5 =freq['Computer_tickets']
resul5
```
