Verificando el tipo fecha datetime64 de la columna date

```
vacunacion.info()
```

Al no estar con el formato especifico se pueden hacer dos cosas

1.- Algunas funciones como read_excel se le puede pasar un parametro parse_date con la lista de columnas a interpretar como fecha
pd.read_excel('/content/archivo.xlsx', parse_dates=['date', 'fecha'])
date y fecha serian los nombres de las columnas del archivo que deseamos que queden como datetime64

2.- Cambiar el formato con la función pd.to_datetime 

Observa que en la siguiente ejecución de info date ya esta como datetime64

```
vacunacion['date'] = pd.to_datetime(vacunacion['date'])
vacunacion.info()
```

Podemos extraer fechas minimas y máximas

```
print("Fecha mínima: ", vacunacion['date'].min())
print("Fecha máxima: ", vacunacion['date'].max())
```

Podemos calcular diferencias entre fechas

```
print('El tiempo reportado en el archiv corresponde a: ', vacunacion['date'].max() - vacunacion['date'].min())
```

Podemos extraer facilmente dia, mes y año de una fecha
