## **Subtema 1.2 Operaciones Matemáticas**

**Resta de arreglos**

```
a = np.array( [10, 20, 30, 40, 50] )
b = np.array( [1, 2, 3, 4, 5] )
resta = a - b
print("Resultado de la resta de arreglos: \n", resta)
```

**Suma de Matrices**

```
rg = np.random.default_rng() 
MatA = rg.integers(20, size=(3, 3))
MatB = rg.integers(5, size=(3, 3))
sumaMat = MatA + MatB
print("MatA: \n", MatA)
print("MatB: \n", MatB)
print("Suma de MatA + MatB: \n", sumaMat)
```

**Arreglo base para las siguientes operaciones**

```
arreglo = np.arange(6)
print("Arreglo: ", arreglo)
```

**Elevar los valores del arreglo a una potencia dada, por ejemplo al cuadrado**

```
cuadrados = arreglo**2
print("Valores del arreglo al cuadrado: ", cuadrados)
```

**Multiplicando el arreglo por un escalar**

```
arrescalar = arreglo * 5
print("Valores del arreglo por un escalar 5: ", arrescalar)
```

**Obteniendo las raices cuadradas de cada uno de los elementos del arreglo**

```
raices = np.sqrt(arreglo)
print("Raices cuadradas calculadas: ", raices)
```

**Multiplicación de Matrices**

```
rg = np.random.default_rng() 
MatA = rg.integers(20, size=(4, 3))
MatB = rg.integers(5, size=(3, 2))
multiplicacionMat = MatA @ MatB
print("MatA: \n", MatA)
print("MatB: \n", MatB)
print("Multiplicación de MatA * MatB: \n", multiplicacionMat)
```

## **Subtema 1.3 Inversa y Transpuesta**

**Matriz Transpuesta**

```
rg = np.random.default_rng() 
MatA = rg.integers(30, size=(4, 3))
transpuesta = np.transpose(MatA)
print("MatA: \n", MatA)
print("Matriz Transpuesta: \n", transpuesta)
```

**Matriz inversa**

```
matA = np.array([[1,2],[3,4]]) 
inversa = np.linalg.inv(matA) 
print ("Matriz A: \n", matA) 
print ("Matriz inversa de A: \n", inversa)
```

**Seleccionando filas con condiciones compuestas (por ejemplo los personajes más altos a 170 cm y con peso mayor a 100 kg)**

**Seleccionando filas y columnas en base a su posición (por ejemplo recuperando los registros del 20 al 30 y las columnas de la 2 a la 6)**

## **Subtema 3.1 Calculando resúmenes con Pandas**

**Aplicando el cálculo del promedio a la columna height para obtener la altura promedio de los personajes de Star Wars**

```
personajesSW = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/M2DCData/characters.csv")
print("La altura promedio de los personajes de star wars es: ", personajesSW["height"].mean())
```

**Aplicando el cálculo del promedio a dos columnas (height y mass) para obtener la altura y peso promedio de los personajes de Star Wars**

```
personajesSW = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/M2DCData/characters.csv")
personajesSW[["height", "mass"]].median()
```

**Aplicando diversos cálculos a varias columnas al mismo tiempo**

```
personajesSW = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/M2DCData/characters.csv")
personajesSW[["height", "mass"]].median()
personajesSW.agg(
        {
            "height": ["min", "max", "median", "skew"],
            "mass": ["min", "max", "median", "mean"],
        }
    )
```
