Leer el archivo vaccination_tweets.csv y dejarlo en un dataframe

Los 5 usuarios con m√°s seguidores

```
masSeguidores = tweets[['user_name', 'user_followers']]
masSeguidores = masSeguidores.sort_values('user_followers')
masSeguidores = masSeguidores.drop_duplicates(subset='user_name')
masSeguidores[:-6:-1]
```

La cantidad promedio de amigos que tienen los usuarios

```
tweets['user_friends'].mean()
```

La cantidad de usuarios por cada fuente de donde hicieron el tweet

```
tweets[['source', 'user_name']].groupby('source').count()
```

El promedio de retweets, sin considerar a aquellos usuarios que no tienen retweetts

```
retweets = tweets[tweets['retweets']>0]
retweets['retweets'].mean()
```

Los 5 primeros usuarios que tienen m√°s de 100 favoritos

```
fav100 = tweets[tweets['favorites']>100]
fav100 = fav100[['user_name', 'favorites']]
fav100.sort_values('favorites', ascending=True, inplace=True)
fav100[:5:]
```
