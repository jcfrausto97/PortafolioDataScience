```python
%%writefile reto.py
import codecs
import pandas as pd
import streamlit as st

st.title("Netflix app")

@st.cache
def load_data(nrows):
    doc = codecs.open('/content/drive/MyDrive/Colab Notebooks/Modulo 5/movies.csv','rU','latin1')
    data = pd.read_csv(doc, nrows=nrows)
    lowercase = lambda x: str(x).lower()
    return data

data_load_state = st.text('Loading cicle nyc data...')
data = load_data(1000)
data_load_state.text("Done! (using st.cache)")


if st.sidebar.checkbox('Mostrar todos los filmes'):
    st.subheader('Todos los filmes')
    st.write(data)


def load_data_byname(name):
   filtered_data_byname = data[data['name'].str.contains(name)]
   return filtered_data_byname

namePeli = st.sidebar.text_input('Nombre de la película: ')
btnPel=st.sidebar.button('Buscar por pelicula')
if (btnPel):
   filterbyname = load_data_byname(namePeli)
   count_row = filterbyname.shape[0]
   st.dataframe(filterbyname)


director = st.sidebar.selectbox("Mostrar todos los directores", data['director'].unique())
st.write(f"Selected Option: {director!r}") 

btnDir=st.sidebar.button('Buscar por Director')
if(btnDir):
   filtered_data_filme = data[data['director'] == director]
   st.dataframe(filtered_data_filme)
```
