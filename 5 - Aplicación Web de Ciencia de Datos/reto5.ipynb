```python
%%writefile employees.py
import codecs
import pandas as pd
import seaborn as sns
import numpy as np
import streamlit as st
import matplotlib.pyplot as plt
import plotly.express as px

st.title("Reto")
st.header("Employees")
st.write("""
# Reto final del modulo 6
""")

@st.cache
def load_data(nrows):
    doc = codecs.open('/content/drive/MyDrive/Colab Notebooks/Modulo 6/Employees.csv','rU','latin1')
    data = pd.read_csv(doc, nrows=nrows)
    lowercase = lambda x: str(x).lower()
    return data

data_load_state = st.text('Loading cicle nyc data...')
data = load_data(500)
data_load_state.text("Done! (using st.cache)")

sidebar = st.sidebar 

if st.sidebar.checkbox('Mostrar DataFrame'):
    st.subheader('DataFrame employees')
    st.write(data)

@st.cache
def load_data(valor):
   filtered_data = data[(data['Employee_ID'].str.contains(valor)) | (data['Unit'].str.contains(valor)) | (data['Hometown'].str.contains(valor))]
   return filtered_data


buscador = st.sidebar.text_input('Buscador (Employee_ID, Unit, Hometown): ')
btn=st.sidebar.button('Buscar')
if (btn):
   filter = (load_data(buscador))
   count_row = filter.shape[0]
   st.dataframe(filter)
   count_row

eduLev = st.sidebar.selectbox("Buscar empleados por nivel educativo", data['Education_Level'].unique())


btnLev=st.sidebar.button('Buscar por Nivel educativo')
if(btnLev):
   filtered_data_Edu = data[data['Education_Level'] == eduLev]
   st.dataframe(filtered_data_Edu)


ciudad = st.sidebar.selectbox("Buscar empleados por Ciudad", data['Hometown'].unique())

btnCiu=st.sidebar.button('Buscar por Ciudad')
if(btnCiu):
   filtered_data_Ciu = data[data['Hometown'] == ciudad]
   st.dataframe(filtered_data_Ciu['Employee_ID'])


searchUnit = st.sidebar.selectbox("Buscar por Unidad", data['Unit'].unique())
btnUnit=st.sidebar.button('Buscar por Unidad')
if(btnUnit):
   filtered_data_Unit = data[data['Unit'] == searchUnit]
   st.dataframe(filtered_data_Unit)

st.header("Empleados agrupados por edad")
#df = pd.DataFrame(data[:1000], columns = ['Age','Employee_ID'])
hist_values = np.histogram(data['Age'], bins=24, range=(0,24))[0]
st.bar_chart(hist_values)

st.header("Frecuencias para las unidades funcionales")
st.line_chart(data['Unit'])


st.header("Índice de deserción por ciudad")
st.write("""
### Lebanon es la ciudad con mayor índice de deserción
""")
df=df = pd.DataFrame(data[:1000], columns = ['Hometown','Attrition_rate'])
fig2 = px.histogram(data_frame=df, x="Hometown", y="Attrition_rate")
st.write(fig2)

st.header("Indice de deserción por edad")
st.write("""
### 30-34 es la edad con mayor índice de deserción
""")
df=df = pd.DataFrame(data[:1000], columns = ['Age','Attrition_rate'])
fig2 = px.histogram(data_frame=df, x="Age", y="Attrition_rate")
st.write(fig2)

st.header("índice de deserción por tiempo de servicio")
st.write("""
### 4-5 años de servicio es el tiempo con mayor índice de deserción
""")
df=df = pd.DataFrame(data[:1000], columns = ['Time_of_service','Attrition_rate'])
fig2 = px.histogram(data_frame=df, x="Time_of_service", y="Attrition_rate")
st.write(fig2)
```
