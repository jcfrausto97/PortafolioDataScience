```python
datos2 = train.merge(train2, left_on='Employee_ID', right_on='Employee_ID', how = 'right')
datos2
```

```python
datos1 = test.merge(test2, left_on='Employee_ID', right_on='Employee_ID', how = 'right')
datos1
```

```python
datos=pd.concat([datos1,datos2], axis = 0)
datos
```

```python
datos.rename(columns={' Computer_OS':'Computer_OS', ' Mobile_OS':'Mobile_OS', ' Mobile_tickets':'Mobile_tickets', ' Computer_tickets':'Computer_tickets'}, inplace=True)
```

```python
#datos = datos.replace({'Computer_OS':'No registradoa'},"No registrado")
#datos = datos.replace({'Mobile_OS':np.nan},"No registrado")
datos["Computer_OS"] = datos["Computer_OS"].replace(np.nan, 'No registrado')
datos["Mobile_OS"] = datos["Mobile_OS"].replace(np.nan, 'No registrado')
```

```python
print(datos.columns.tolist())
```

### El porcentaje de personas que usan Windows / Mac / Linux / no registrado como su sistema operativo en su computadora 

```python
freq = datos['Computer_OS'].value_counts() 
total = datos['Computer_OS'].count()
resul = (freq/total)*100
resul
```
