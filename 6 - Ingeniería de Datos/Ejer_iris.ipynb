```python
# Tratameinto de datos
# ====================================================================
import numpy as np
import pandas as pd
import statsmodels.api as sm

# Preprocesado y modelado
# ====================================================================
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import scale

# Configuración warnings
# ====================================================================
import warnings 
warnings.filterwarnings('ignore')
```

```python
iris = sm.datasets.get_rdataset("iris", "datasets")
datos = iris.data
```

```python
datos
```

```python
#es muy importante que nuestros datos tengan media 0
print("Media de cada variable")
datos.mean()
```

```python
#desviación estándar debe ser de 1
print("Desviación estándar de cada variable")
datos.std()
```

```python
#Lo primero que se hace es asignar a un objeto llamado “escalar” la función de Python que se encarga de estandarizar los datos (StandardScaler())
escalar = StandardScaler()
```

```python
#Dividimos nuestros datos en X y Y
# suponiendo que la variable Species en un problema de clasificación sería la variable que quisiéramos predecir, es decir, quisiéramos saber dado que características adicionales se puede clasificar una flor en sus diferentes especies.   
#para iris
features = ['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']
# Separating out the features
x = datos.loc[:, features]
# Separating out the target 
y = datos.loc[:,['Species']]
```

```python
x.head()
```

```python
y.head()
```

```python
 # .fit es la función encargada de calcular la media y desviación estándar para posteriormente utilizar esos valores
datos_escalados = escalar.fit(x)
# con .transform transforma los valores reales a valores estandarizados.
datos_escalados = escalar.transform(x)
datos_escalados
```

```python
#Empezamos a aplicar PCA
#Lo primero que se hace es construir el modelo de reducción de dimensionalidad
#es necesario darle ciertos parámetros para su correcta implementación, es por eso que nosotros le vamos a indicar cuántos componentes construir en primera instancia.
#no hay una regla definida para determinar inmediatamente cuantos componentes principales construir.
#Utilizaremos la opción de indica que el número de componentes principales que se pueden calcular es como máximo el menor valor entre n-1 o p, por lo tanto, 
#nosotros n-1 = 150 – 1 = 149 y p = 4, por lo tanto, 4 componentes principales son los que vamos a calcular.   
pca_model = PCA(n_components = 4)
```

```python
#al modelo construido (pca_model) le damos nuestros datos de entrada X 
#originales para que haga la transformación. 
pca_model.fit(datos_escalados)
```

```python
#muestra un arreglo de valores con la carga de cada una de nuestras variables originales en relación con los componentes principales, 
pca_model.components_
```

```python
#Vamos a utilizar como datos el arreglo de valores de la carga, vamos a tener como columnas los nombres de las columnas de los datos originales y finalmente 
#nuestros renglones van a estar asociados a los componentes principales creados, 
pd.DataFrame(
    data    =  pca_model.components_,
    columns = x.columns,
    index   = ['PC1', 'PC2', 'PC3', 'PC4'] 
)
```

```python
#Hay que utilizar la matriz de carga de las variables y multiplicarlo por los valores originales de nuestras variables de entrada (esto corresponde a una multiplicación de matrices punto a punto).  
proyecciones = np.dot(pca_model.components_, scale(x).T)
#Posteriormente asignamos los nombres correspondientes a cada componente principal.  
proyecciones = pd.DataFrame(proyecciones, index = ['PC1', 'PC2', 'PC3', 'PC4'])
#Para finalmente, lo transponemos para que quede en el formato de nuestros datos originales:  
proyecciones = proyecciones.transpose().set_index(x.index)
proyecciones
```

```python
#Seleccionamos 2 componentes principales para nuestro primer conjunto de datos
conjunto_1 = proyecciones [["PC1", "PC2"]] 
conjunto_1.head()
```

```python
conjunto_2 = proyecciones [["PC1", "PC2", "PC3"]] 
conjunto_2.head()
```

```python
#importante recordar que tenemos una variable y que no consideramos para el cálculo de componentes principales y sería necesario incluirla 
#nuevamente en los conjuntos de datos finales para poder ser utilizada en otros modelos más adelante
conjunto_1 = pd.concat([conjunto_1, datos [['Species']]], axis = 1)
conjunto_1
```

```python
conjunto_2 = pd.concat([conjunto_2, datos [['Species']]], axis = 1)
conjunto_2
```
