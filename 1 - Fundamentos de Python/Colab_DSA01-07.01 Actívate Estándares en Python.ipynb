Ejercicio 4

```
list = [1, 2, 3] # This is an example of a single line comment.
for n in list:
# This is an example of a block comment.
# This comment belongs to the if.
  if n == 0:
    print('N is 0')
  else:
    print('N is not 0')

# This is also an example of a single line comment.
dictionary = {'key':'value'}
print(dictionary['key'])
```

Ejercicio 5

```
list = [1, 2, 3] # Esto es un ejemplo de un comentario dentro de una línea. 
for n in list:
  # Esto es un ejemplo de un comentario en bloque.
  # El comentario pertenece al estatuto if.
  if n == 0:
    print('N is 0')
  else:
    print('N is not 0')

# Esto es otro ejemplo de un comentario en bloque.
# El comentario pertenece a la asignación de un nuevo diccionario.
dictionary = {'key':'value'}
print(dictionary['key'])
```
