***Ejemplo***

Contar el número de renglones que contiene un archivo

```
file_handler = open('/content/sample_data/prueba.txt', 'r')
count = 0
for line in file_handler:
  count = count + 1
print('Total number of lines:', count)
file_handler.close()
```

***Ejercicios***

1. Calcular la suma de los números que se encuentran en el archivo. El resultado debe ser igual a 130. 









```
file_handler = open('/content/sample_data/prueba.txt')
x = 0
for line in file_handler :
    x = x + int(line)
    print(x)

file_handler.close()
```

2. Calcular el promedio de los números que se encuentran en el archivo. El resultado debe ser igual a 26. 



```
file_handler = open('/content/sample_data/prueba.txt')

x = 0
y = 0

for line in file_handler :
    x = x + int(line)
    y = y + 1

print( x )
print( x / y)

file_handler.close()
```
