##TEMA 2. Definición de funciones

***Ejemplo***

Calcular el área de un triángulo.
<br>Entradas: base y altura
<br>Salida: área del triángulo

```
def triangle_area(b, h):
  return b * h / 2

a = triangle_area(2.0, 5.0)
print('Area:', a)
```
